### Problem 10

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]
````

````python
# YOUR CHANGES HERE

q1010 = [15]
for i in range (10):
    if q1010[i] % 2 == 0:
        q1010.append(q1010[i]//2)
    else:
        q1010.append(int(3*q1010[i] + 1))

plt.figure(figsize=(8,5))
plt.plot(q1010, 'Purple', marker='s', linewidth=2, markersize=6, markerfacecolor="#00FF77", markeredgecolor="#0a8c93", label =f"Starting from {q1010[0]}")
plt.title('Collatz sequences - 10 entries')
plt.xlabel('Step')
plt.ylabel('Value')
plt.xticks(range(len(q1010)))
plt.grid(True, alpha=0.25)
plt.legend(loc="upper left")
````
